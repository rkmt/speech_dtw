```python
sr = 16000

def show_wave(w, sr=sr):
    plt.figure(figsize=(14, 5))
    librosa.display.waveplot(w, sr=sr)
    
def mcepalpha(fs):
    """
    Determine frequency warping parameter from fs (simplest implementation)
    See https://bitbucket.org/happyalu/mcep_alpha_calc/ for details.
    """
    if fs == 16000:
        return 0.41
    elif fs == 22050:
        return 0.455
    elif fs == 44100:
        return 0.544
    elif fs == 48000:
        return 0.554
    else:
        raise NotImplementedError


def decompose(x, fs=sr, frame_period=10, f0_floor=40.0, f0_ceil=1500.0):
    """
    Decompose speech signal into f0, spectral envelope and aperiodicity using WORLD
    """
    f0, timeaxis = pw.harvest(x, fs, frame_period=frame_period, f0_floor=f0_floor, f0_ceil=f0_ceil)
    sp = pw.cheaptrick(x, f0, timeaxis, fs)
    ap = pw.d4c(x, f0, timeaxis, fs)
    return f0, timeaxis, sp, ap


def w2sptk(w, dim=32, frame_period=10, fs=sr, f0_floor=40.0, f0_ceil=1500.0):
    w = w.astype(np.double)
    hop_length = int(fs * frame_period * 0.001)
    fftlen = pw.get_cheaptrick_fft_size(fs)
    #print("hop_length", hop_length, "fftlen", fftlen)
    f0, timeaxis, sp, ap = decompose(w, fs, frame_period=frame_period, f0_floor=f0_floor, f0_ceil=f0_ceil)
    alpha = mcepalpha(fs)
    mc = np.apply_along_axis(pysptk.sp2mc, 1, sp, dim-1, alpha)
    return mc, f0, sp
```

## 波形読み込み

```python
# wave1 読み込み
w1, _ = librosa.load("jpn.wav", sr=sr)
show_wave(w1)
ipd.Audio(w1, rate=sr)
```

```python
# wave2 読み込み
w2, _ = librosa.load("eng.wav", sr=sr)
show_wave(w2)
ipd.Audio(w2, rate=sr)
```

## DTW による音響特徴量(MFCC)の対応づけ

```python
mfcc1, f0_1, sp1 = w2sptk(w1)
mfcc2, f0_2, sp1 = w2sptk(w2)
print(mfcc1.shape, mfcc2.shape)



alignment = dtw.dtw(mfcc1, mfcc2, keep_internals=True)
wt = dtw.warp(alignment, index_reference=True)
alignment.plot('alignment')
```

## wave1 の区間　を wave2 の区間に当てはめてみる

```python
def align_time(w1, mfcc1, wt, pos1):
    '''
    w1 上の pos1 は w2上の pos2
    '''
    numframe = len(w1) / mfcc1.shape[0] # 特徴量あたりの波形長さ
    return (int)(numframe * wt[(int)(pos1/numframe)])
```

```python
# w1 の部分を切り出す (from_time, to_time 秒)
from_time = 2.5
to_time = from_time + 0.8

w1clip = w1[(int)(from_time*sr):(int)(to_time*sr)]
show_wave(w1clip)
ipd.Audio(w1clip, rate=sr)
```

```python
# w1の from_time, to_time に相当するw2の部分を切り出す
from_pos = align_time(w1, mfcc1, wt, (int)(sr*from_time))
to_pos = align_time(w1, mfcc1, wt, (int)(sr*to_time))
print(from_pos, to_pos)

w2clip = w2[from_pos:to_pos]
show_wave(w2clip)
ipd.Audio(w2clip, rate=sr)
```
